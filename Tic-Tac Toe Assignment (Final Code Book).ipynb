```python

from IPython.display import clear_output
from random import randint

# Creating a display board for Tic Tac
def display_board(board):
    clear_output()
    print(board[7]+'|'+ board[8]+'|'+ board[9])
    print(board[4]+'|'+ board[5]+'|'+ board[6])
    print(board[1]+'|'+ board[2]+'|'+ board[3])

# Choosing Markers for the Players
mark = {}

def entry():
    result = input ('Player 1: Enter your marker choice: ').upper()
    return result

# Choose the player that goes first
def choose_player():
    if randint(0,1) == 0:
        return 'Player 1'
    else:
        return 'Player 2'

# Asking the position of the marker
def pos(board,player):
    position = 0
    a = input('Enter your position on the board (1-9): ')
    position = int (a)
    while not position in list(range(1,10)) or not board[position] == '':
        clear_output()
        display_board(board)
        b = input('Enter your position on the board (1-9): ')
        position = int(b)
    else:
        return position

def place_marker_board(board,position, mark, player):
    board[position] = mark[player]

def win_check (board,mark,player):
    
    mark_base = []
    mark_base.append(mark[player])
    
    return ( list(board[7] + board[8] + board[9]) == mark_base*3 or 
             list(board[4] + board[5] + board[6]) == mark_base*3 or
             list(board[1] + board[2] + board[3]) == mark_base*3 or
             list(board[7] + board[5] + board[3]) == mark_base*3 or
             list(board[1] + board[5] + board[9]) == mark_base*3 or
             list(board[1] + board[4] + board[7]) == mark_base*3 or
             list(board[3] + board[6] + board[9]) == mark_base*3 or
             list(board[8] + board[5] + board[2]) == mark_base*3 )

def full_board_check(board):
    for i in range(1,10):
        if board[i] == '':
            return False
    else:
        return True

def replay():
    return input('Do you want to play again? Enter Yes or No: ').lower().startswith('y')
    
# Program Starts

print('Welcome to Tic Tac Toe!')

while True:
    board = ['']*10  # Creating the list for board display and capturing the marker position on the board

    mark = {} # Choosing Markers for the Players
    
    marker = entry()
        
    while not (marker == 'X' or marker == 'O'):
        clear_output()
        print ('Please choose X or O')
        marker = entry()
    else:
        clear_output()

    if marker == 'X':
        mark ['Player 1'] = 'X'
        mark ['Player 2'] = 'O'
        print ('You have chosen X as your marker\nYour Partner is assigned O as a marker')
    elif marker == 'O':
        mark ['Player 1'] = 'O'
        mark ['Player 2'] = 'X'
        print ('You have chosen O as your marker\nYour Partner is assigned X as a marker')
        
    turn = choose_player()
    print (turn + ' will go first')
    
    play_game = input('Are you ready to play (Y/N): ').lower()
    
    
    if play_game == 'y':
        game_on = True
    else:
        game_on = False
        
    while game_on:
        
        if turn == 'Player 1':
            display_board(board)
            position = pos(board,turn)
            place_marker_board(board,position, mark, turn)
            
            if win_check(board,mark,turn):
                display_board(board)
                print(f'{turn} Congratulations! You have won the game!')
                game_on = False
            
            elif full_board_check(board):
                display_board(board)
                print('The game is a draw!')
                break

            else:
                turn = 'Player 2'
        
        else:
            display_board(board)
            position = pos(board,turn)
            place_marker_board(board,position, mark, turn)
            
            if win_check(board,mark,turn):
                display_board(board)
                print(f'{turn} Congratulations! You have won the game!')
                game_on = False
            
            elif full_board_check(board):
                display_board(board)
                print('The game is a draw!')
                break
            
            else:
                turn = 'Player 1'
    
    if not replay():
        break
    else:
        continue
```
