```python
import matplotlib.pyplot as plt

import numpy as np
```

```python
x = np.linspace(0,10,50)
y = x**2
```

```python
fig = plt.figure()
axes = fig.add_axes([0.5,0.1,0.8,0.8])
axes_1 = fig.add_axes([0.6,0.5,0.3,0.3])
axes.plot(x,y)
axes_1.plot(y,x)
axes.set_title("Large Plot")
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
plt.show()
```

```python
fig_1, axes_2 = plt.subplots(nrows=3,ncols=2,figsize=(8,8),dpi=300)
plt.tight_layout()
axes_2[1,1].set_title('Test')
plt.show()
```
